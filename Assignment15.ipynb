Q1. What is Flask framework? What are the advantages of Flask framework?

Ans: Flask is a micro web framework written in Python. It is classified as a microframework because it does not require            particuler tools or libraries. It has no database abstraction layer, form validation, or any other components where
     pre-existing third party libraries provide common functions.
     
     Advantages:
     1. Scalable
     2. Flexible
     3. Easy to negotiate
     4. Lightweight
     5. Documentation

Q2. Create a simple Flask assignment to display 'Hello World!!'. Attach the screenshot of the output in Jupytar Notebook.

```python
#Ans:
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app router?

Ans: App routing means mapping the URLs to a specific function that will handle the logic for that URL. 

     App routing is used to map the specific URL with the specific function that is intended to perform some task. The             routing technique helps a user remember application URLs and is useful to access the desired page directly without             having to navigate from the home page.

Q5. What function is used in Flask for URL Building? Write a python code to demonstrate the working of the url_for() function.

Ans: url_for() function is used to build a URL in Flask.


```python
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route("/admin")
def hello_admin():
    return "Hello Admin"
@app.route('/guest/<guest>')
def hello_guest(guest):
    return 'Hello %s as Guest' % guest
@app.route('/user/<name>')
def hello_user(name):                              # the function hello_user(name) which accept a value to its argument form 
    if name=='admin':                              # the url. The hello_user() function check if an argument received matches
        return redirect(url_for('hello_admin'))    # 'admin' or not. If not matches, the application is redirected to the hello 
    else:                                          # _admin function using url_for(). Otherwise to the hello_guest function
        return redirect(url_for('hello_guest',guest=name))    # passing the received argument as guest parameter to it. 
if __name__=='__main__':
    app.run(debug=True)                    
```
